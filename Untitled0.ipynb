```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
n_estimators_list=[100,200,300,400,500]
max_depth_list=[5,10,15,20]
results=[]
for n_trees in n_estimators_list:
  for depth in max_depth_list:
    rf=RandomForestClassifier(n_estimators=n_trees,max_depth=depth,random_state=42,class_weight='balanced')
    rf.fit(X_train,Y_train)
    y_pred=rf.predict(X_test)
    accuracy=accuracy_score(Y_test,y_pred)
    recall=recall_score(Y_test,y_pred)
    precision=precision_score(Y_test,y_pred)
    f1=f1_score(Y_test,y_pred)
    results.append ({'Trees': n_trees, 'Depth':depth, 'Accuracy':accuracy,'Recall':recall,'precision': precision,'F1 Score':f1})
result_df=pd.DataFrame(results)
result_df=result_df.sort_values(by=['Recall','Accuracy'],ascending=False)
print(result_df.head(20))
```

**CUSTOMER SEGMENTATION**

```python
# Recalculate churn probabilities for the entire dataset X
churn_prob_full = rf_tuned.predict_proba(X)[:, 1]

segmentation_data=pd.DataFrame({
'Tenure Months': X['Tenure Months'],
'Monthly Charges':X['Monthly Charges'],
'Total Charges':X['Total Charges'],
'Churn Probability':churn_prob_full})
```

```python
segmentation_data
```

**implementattion of kmeans**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaled_data=scaler.fit_transform(segmentation_data)
print(scaled_data[:5])
```

```python
from sklearn.cluster import KMeans
wcss=[]
# Ensure scaled_data does not contain NaNs before fitting KMeans
# This step handles potential NaNs if upstream data processing was not fully consistent.
# Filtering NaNs here ensures KMeans can proceed.
cleaned_scaled_data = scaled_data[~np.isnan(scaled_data).any(axis=1)]

# Now loop for KMeans
for k in range(1,16):
  kmeans=KMeans(n_clusters=k, random_state=42, n_init='auto')
  kmeans.fit(cleaned_scaled_data)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,6))
plt.plot(range(1,16),wcss,marker='o')
plt.xlabel("Number of cluster k ")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
```

```python
clusters=kmeans.fit_predict(scaled_data)
```

```python
segmentation_data
```

```python
segmentation_data['Cluster']= clusters
```
